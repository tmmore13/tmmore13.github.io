```python
import matplotlib.pyplot as plt
import numpy as np

# Create x-axis data (from 0 to 2*pi)
x = np.linspace(0, 2 * np.pi, 200)

# Create y-axis data for the two functions
y1 = np.sin(x) # Sine function
y2 = np.cos(x) # Cosine function

# Create the plot
plt.figure(figsize=(10, 6)) # Sets the figure size

# Plot the first function (sine) in blue
plt.plot(x, y1, label='sin(x)', color='blue')

# Plot the second function (cosine) in red
plt.plot(x, y2, label='cos(x)', color='red')

# Add titles and labels for clarity
plt.title('Plot of Sine and Cosine Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() # Displays the labels for each line
plt.grid(True) # Adds a grid to the background

# Show the plot
plt.show()
```
