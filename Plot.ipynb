```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Read the Excel file into a pandas DataFrame
file_path = 'Cancer_Cases_deaths_2025.xlsx'
df = pd.read_excel(file_path)

# --- DATA CLEANING ---
# The error is caused by the row at index 10 which contains '-'.
# We will remove any rows where 'Common Type of Cancers' is '-'
df = df[df['Common Type of Cancers'] != '-']

# Also, let's remove the summary row "Cancer of Any Site" to compare only the specific types
df = df[df['Common Type of Cancers'] != 'Cancer of Any Site']


# --- PLOT 1: Estimated New Cases ---
print("Plotting Estimated New Cases...")
plt.figure(figsize=(10, 8)) # Adjusted size for better readability

# A horizontal bar chart is better here because the cancer type names are long
sns.barplot(
    x='Estimated New Cases 2025',
    y='Common Type of Cancers',
    data=df.sort_values('Estimated New Cases 2025', ascending=False), # Sort for a cleaner look
    palette='viridis'
)

# Update titles and labels to match the data
plt.title('Estimated New Cancer Cases in 2025', fontsize=16)
plt.xlabel('Estimated Number of New Cases', fontsize=12)
plt.ylabel('Type of Cancer', fontsize=12)

plt.tight_layout()
plt.show()


# --- PLOT 2: Estimated Deaths ---
print("\nPlotting Estimated Deaths...")
plt.figure(figsize=(10, 8))

sns.barplot(
    x='Estimated Deaths 2025',
    y='Common Type of Cancers',
    data=df.sort_values('Estimated Deaths 2025', ascending=False), # Sort for a cleaner look
    palette='plasma'
)

# Update titles and labels
plt.title('Estimated Cancer Deaths in 2025', fontsize=16)
plt.xlabel('Estimated Number of Deaths', fontsize=12)
plt.ylabel('Type of Cancer', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read and clean the Excel file data
file_path = 'Cancer_Cases_deaths_2025.xlsx'
df = pd.read_excel(file_path)

# --- DATA CLEANING ---
# Remove the row with hyphens and the summary row
df = df[df['Common Type of Cancers'] != '-']
df = df[df['Common Type of Cancers'] != 'Cancer of Any Site']


# --- PLOT 1: Pie Chart for Estimated New Cases ---
print("Plotting Pie Chart for Estimated New Cases...")
plt.figure(figsize=(10, 8)) # Set the figure size

plt.pie(
    df['Estimated New Cases 2025'],
    labels=df['Common Type of Cancers'],
    autopct='%1.1f%%',  # Format to show percentages with one decimal place
    startangle=140       # Rotate the pie chart to start at a nice angle
)

plt.title('Proportion of Estimated New Cancer Cases in 2025', fontsize=16)
plt.axis('equal')  # Ensures that the pie is drawn as a circle.
plt.show()


# --- PLOT 2: Pie Chart for Estimated Deaths ---
print("\nPlotting Pie Chart for Estimated Deaths...")
plt.figure(figsize=(10, 8))

plt.pie(
    df['Estimated Deaths 2025'],
    labels=df['Common Type of Cancers'],
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Proportion of Estimated Cancer Deaths in 2025', fontsize=16)
plt.axis('equal')  # Ensures that the pie is drawn as a circle.
plt.show()
```
