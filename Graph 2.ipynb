```python
import matplotlib.pyplot as plt

# --- Data for the Bar Chart ---
# These are the labels for each bar on the x-axis.
categories = ['Group A', 'Group B', 'Group C', 'Group D']

# These are the numerical values that determine the height of each bar.
values = [15, 24, 12, 19]

# --- Creating the Plot ---
# You can set the size of your plot figure (optional).
plt.figure(figsize=(8, 6))

# The core command to create the bar chart.
# We also add different colors to each bar for better visualization.
plt.bar(categories, values, color=['skyblue', 'lightgreen', 'salmon', 'gold'])

# --- Adding Labels and Title ---
# Add a title to the chart.

# Add a label to the x-axis.
plt.xlabel('Group Name')

# Add a label to the y-axis.
plt.ylabel('Scores')

# --- Displaying the Plot ---
# This command displays the final chart.
plt.show()
```
